```python
import pandas as pd
```

```python
lens = []
names = []
```

```python
for name, g in pd.DataFrame(a).groupby(0):
    lens.append(g[1].max())
    names.append(name)
```

```python
names = np.array(names)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(lens, bins=100)
plt.xlabel('Episode Length')
plt.savefig('hist.png')
```

```python
import numpy as np

u, c = np.unique(lens, return_counts=True)
reps = u[c>1]
```

```python
raw_data = torch.load('/data1/s3092593/qbert_replays/DQN_modern/Qbert/0/model_50000000_new/train/data.pt')
```

```python
raw_data[0]
```

```python
for rep in reps:
    print(names[lens == rep])
    acts_comps = []
    for eid in names[lens == rep]:
        for episode, rewards, actions in raw_data:
            if episode == eid:
                acts_comps.append(actions)
    for arr in acts_comps[1: ]:
        if arr == acts_comps[0]:
            print('mm')
```
